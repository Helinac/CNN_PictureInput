```python
import pandas as pd
import mplfinance as mpf
```

```python
my_color = mpf.make_marketcolors(ohlc='white',volume='white')
my_style = mpf.make_mpf_style(marketcolors=my_color,
                              mavcolors = ['white'],
                              figcolor='(0,0,0)')
my_config = dict(ohlc_ticksize = 0.5,
                     ohlc_linewidth = 3,
                     volume_width = 0,
                     volume_linewidth = 3,
                     line_width = 2)
```

```python
df = pd.read_csv('AAPL.csv',index_col = 'Date')

df.index = pd.to_datetime(df.index)

df
```

```python
AAPL1 = df.loc['2009-01-30' : '2009-02-27']
```

```python
ap = mpf.make_addplot(AAPL1['20-day Avg.'],type='line',color = "white")
```

```python
mpf.plot(AAPL1,
        style=my_style,
        type='ohlc',
        volume=True,
        addplot=ap,
        figsize=(3,3),
        axisoff=True,
        tight_layout=True,
        scale_padding=1,
        update_width_config = my_config
        )
```
